```python
attempt_to_rape_df = pd.read_csv('D:\Downloads\A Portfolio Projects\Python EDA projects\Women Harassment Bangladesh\Dataset\Attempt-to-Rape_2014-2021.csv')
suicide_after_rape_df = pd.read_csv('D:\Downloads\A Portfolio Projects\Python EDA projects\Women Harassment Bangladesh\Dataset\Rape_CommittedSuicideAfterBeingRaped_2007-2021.csv')
gang_rape_df = pd.read_csv('D:\Downloads\A Portfolio Projects\Python EDA projects\Women Harassment Bangladesh\Dataset\Rape_GangRape_2007-2021.csv')
killed_after_rape_df = pd.read_csv('D:\Downloads\A Portfolio Projects\Python EDA projects\Women Harassment Bangladesh\Dataset\Rape_KilledAfterRape_2007-2021.csv')
total_rape_df = pd.read_csv('D:\Downloads\A Portfolio Projects\Python EDA projects\Women Harassment Bangladesh\Dataset\Rape_Total_2001-2021.csv')
stalking_df = pd.read_csv('D:\Downloads\A Portfolio Projects\Python EDA projects\Women Harassment Bangladesh\Dataset\Stalking_2011-2021.csv')
```

### Victims of Stalking for last 11 years.(Only Reported Case)


```python
plt.figure(figsize=(14,3))
stalking_mean_victims = round(stalking_df.victims.mean())
sns.set_style('dark')
plt.subplot(121).set_title('Number of stalking victims over the years from (2011-2021)', fontdict={'fontsize' : 9})
colors = ['r' if victim >= stalking_mean_victims else 'g' for victim in stalking_df.sort_values(by='year', ascending=True).victims]
sns.barplot(data=stalking_df, x='year', y='victims', palette=colors).axhline(stalking_mean_victims, c='grey', ls='--', label=f'average: {stalking_mean_victims}')
plt.legend()
sns.set_style('darkgrid')
plt.subplot(122).set_title('Number of protesting victims over the years from (2011-2021)', fontdict={'fontsize' : 9})
sns.lineplot(data=stalking_df, x='year', y='protesting_female_attacked', color='pink', marker='o', ls='--', label='female')
sns.lineplot(data=stalking_df, x='year', y='protesting_male_attacked', color='blue', marker='s', ls='-', label='male')

# Add data labels for female victims
for i in range(len(stalking_df)):
    plt.text(stalking_df['year'][i], stalking_df['protesting_female_attacked'][i], 
    stalking_df['protesting_female_attacked'][i], color='black', ha='center', va='bottom')
    
# Add data labels for male victims
for i in range(len(stalking_df)):
    plt.text(stalking_df['year'][i], stalking_df['protesting_male_attacked'][i], 
    stalking_df['protesting_male_attacked'][i], color='black', ha='center', va='bottom')
    
plt.xticks(stalking_df['year'])
plt.ylabel('No of protesting victims')
plt.yticks([])
plt.legend()
plt.show()
```

#### <b>Analysis:</b> 
The provided graphs offer a comprehensive overview of stalking incidents and the repercussions faced by those protesting such behaviors over an eleven-year span. Here is a refined interpretation:

<b>Stalking Trends:</b>  The left graph reveals an initial surge in stalking cases in 2011, followed by a gradual decrease. Notably, the frequency of incidents exceeded the eleven-year average between 2011 and 2013, indicating a critical period for stalking occurrences.

<b>Protestor Victimization:</b>  The right graph indicates a decline in attacks against anti-stalking protestors. However, a consistent pattern emerges: male protestors are disproportionately targeted compared to females, suggesting a gender-specific risk factor in these retaliatory acts.

<b>Prevalence of Rape:</b>  The data suggests that actual rape incidents occur at a higher rate than attempts, which is a concerning indication of the prevalence of this crime.
In addition to these insights, it is crucial to consider the role of legal and social support systems in encouraging reporting, as well as the need for comprehensive education and prevention programs to address and reduce the occurrence of such heinous crimes.

### Trend Analysis of Overall Rape in two decades of Bangladesh from 2001 to 2022

```python
plt.figure(figsize=(14,4))
plt.title('Trend Analysis of Overall Rape in two decades of Bangladesh from 2001 to 2022')
sns.set_style('darkgrid')
sns.lineplot(data=total_rape_df, x='Years', y='Total_victims', marker='s', ls='-', color='red')
for i in range(0,total_rape_df.shape[0]):
    plt.text(x=total_rape_df['Years'][i], y=total_rape_df['Total_victims'][i], s=total_rape_df['Total_victims'][i],color='black', ha='center', va='bottom')
plt.xticks(total_rape_df.Years)
plt.show()
```

#### Observations from the Graph
<b>Cyclical Trends:</b> The graph shows that rape rates go up and down over time. These ups and downs might be because of changes in how society deals with the issue or new preventive measures being put in place.

<b>Effective Control Period (2006-2018):</b> During these years, the rates were stable, meaning there weren't many big changes up or down. This suggests that whatever strategies or measures were being used to prevent rape were working well.

<b>Peak in 2022:</b> There was a significant increase in rape rates in 2022. However, based on past trends, we expect this rate to go down again after the peak. This could be because when rates peak, more attention and resources are usually directed towards prevention.

### Ratio of Children & Women (Total Rape VS Gang Rape)

```python
plt.figure(figsize=(18,6))
sns.set_style('darkgrid')
plt.subplot(121).set_title('Rape Over the years on women and children')
sns.lineplot(data=total_rape_df, x='Years', y='women', color='blue', marker='s', label='women')
sns.lineplot(data=total_rape_df, x='Years', y='children', color='red', marker='o', label='children')
sns.lineplot(data=total_rape_df, x='Years', y='Unidentified_age', color='green', label='Unindentified age')
plt.xticks(range(min(total_rape_df.Years),max(total_rape_df.Years)+2,2))
plt.ylabel('Number')
plt.legend()
plt.subplot(122).set_title('Gang Rape over the years on women and children')
sns.lineplot(data=gang_rape_df, x='Years', y='women', color='blue', marker='s', label='women' )
sns.lineplot(data=gang_rape_df, x='Years', y='children', color='red', marker='o',label='children' )
sns.lineplot(data=gang_rape_df, x='Years', y='Unidentified_age', color='green', label='Unidentified age')
plt.ylabel('Number')
plt.xticks(range(gang_rape_df.Years.min(), gang_rape_df.Years.max()+2,2))
plt.legend()
plt.suptitle('Rape vs Gang Rape')
plt.show()
```

#### Analysis:
The graphs compare rape and gang rape incidents involving women and children from 2001 to 2021. Here are the key points:

<b>Rape Trends:</b> Historically, the rape rate for women was higher.
From 2007, the rape rate for children abruptly surpassed that of women. This change might be due to better reporting or societal factors. An additional factor could be that children may be more afraid and reluctant to report or disclose incidents, potentially reinforcing as a support for perpetrators.

<b>Gang Rape Trends:</b> Gang rape incidents involving women have consistently been higher than those involving children, but the difference is relatively small.

These observations suggest:

The sudden increase in child victims from 2007 needs further investigation.
The similar rates of gang rape for women and children indicate the need for targeted interventions.

### Total Rape Compare to Total Death After Rape


```python
rape_from_2007 = total_rape_df[total_rape_df['Years']>=2007]
rape_from_2007['Total_death'] = suicide_after_rape_df.Total_victims + killed_after_rape_df.Total_victims
```

```python
rape_from_2007
```

```python
plt.figure(figsize=(10,4))
sns.set_style('darkgrid')
sns.barplot(data=rape_from_2007, x='Years', y='Total_victims', color='Green', label='Victims', alpha=0.8)
sns.barplot(data=rape_from_2007, x='Years', y='Total_death', color='red', label='Death')
plt.legend()
plt.show()
```

#### Analysis:
The total number of deaths after rape is substantially lower than the total number of rapes, as shown in the graph above. There could be a number of causes:
- Following a suicide, people are less likely to register a complaint (as suicides are also included in total deaths).

- Police were unable to determine the true motive behind the murder after rapping. Because of this, it is listed as a typical murder.
